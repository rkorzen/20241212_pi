```python
import csv

with open("dane/plik.csv") as f:
    reader = csv.DictReader(f, delimiter=";", quotechar='"')

    for row in reader:
        print(row)
```

```python
dane = [
    {"A": 1, "B": 2, "C": 3},
    {"A": 4, "B": 5, "C": 6},
    {"A": 7, "B": "ala;kot", "C": 9},
]

with open("dane/plik2.csv", "w") as f:
    writer = csv.DictWriter(f, fieldnames=["A", "B", "C"], delimiter=",", quotechar='"')
    writer.writeheader()
    writer.writerows(dane)
```

```python
dane = [
    ["A", "B", "C"],
    [1, 2, 3],
    [4, 5, 6],
    [7, "ala;kot", 9],
]

with open("dane/plik3.csv", "w") as f:
    writer = csv.writer(f, delimiter=",", quotechar='"')
    # writer.writeheader()
    writer.writerows(dane)
```
